```python
import numpy as np
import pandas as pd
import ast
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
Movies = Movies[['movie_id','title','overview','genres','keywords','cast','crew']]
Movies.head()
```

```python
Movies.isnull().sum()
```

```python
Movies.dropna(inplace=True)
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
Movies['genres'] = Movies['genres'].apply(convert)
Movies.head()
```

```python
Movies['keywords'] = Movies['keywords'].apply(convert)
Movies.head()
```

```python
def convert2(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter != 3 :
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
Movies['cast'] = Movies['cast'].apply(convert2)
Movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
Movies['crew'] = Movies['crew'].apply(fetch_director)
Movies.head()
```

```python
Movies['overview'][0]
```

```python
Movies['overview'] = Movies['overview'].apply(lambda x:x.split())
```

```python
Movies.head()
```

```python
def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(' ', ''))
    return L1
```

```python
Movies['cast'] = Movies['cast'].apply(remove_space)
Movies['crew'] = Movies['crew'].apply(remove_space)
Movies['keywords'] = Movies['keywords'].apply(remove_space)
Movies['genres'] = Movies['genres'].apply(remove_space)
```

```python
Movies.head()
```

```python
Movies['tags'] = Movies['overview'] + Movies['genres'] + Movies['keywords'] + Movies['cast'] + Movies['crew']
```

```python
Movies.head()
```

```python
Movies_2 = Movies.drop(columns=['overview', 'genres', 'keywords', 'cast', 'crew'])
Movies_2.head()
```

```python
Movies_2['tags'] = Movies_2['tags'].apply(lambda x: " ".join(x))
Movies_2.head()
```

```python
Movies_2['tags'] = Movies_2['tags'].apply(lambda x:x.lower())
Movies_2.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y) 
```

```python
Movies_2['tags'] = Movies_2['tags'].apply(stem)
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(Movies_2['tags']).toarray()
```

```python
vector.shape
```

```python
vector[0]
```

```python
Movies_2['tags'].head()
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
similarity[0].shape 
```

```python
def recommender(movie):
    movie_index = Movies_2[Movies_2['title'] == movie].index[0]
    distance = similarity[movie_index]
    movies_lists = sorted(list(enumerate(distance)), reverse=True, key=lambda x:x[1])[0:6]
    
    for i in movies_lists:
        print(Movies_2.iloc[i[0]].title)
```

```python
recommender('Batman Begins')
```
